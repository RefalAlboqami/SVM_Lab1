```python
def limits(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1

    upper = Q3 + 1.5*IQR
    lower = Q1 - 1.5*IQR

    print("Upper Bound:",upper)   
    print("Lower Bound:",lower)
    return upper, lower


print("sepal width (cm)")
upper1, lower1 = limits(df['sepal width (cm)'])
```

# Plot a decision boundary for a given model

```python
X = iris.data[:, :2]
y = iris.target

logreg = LogisticRegression(C=1e5)
logreg.fit(X, y)

def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    x1, y1 = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    
    Z = model.predict(np.c_[x1.ravel(), y1.ravel()])
    Z = Z.reshape(x1.shape)
    
    plt.contourf(x1, y1, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.show()

plot_decision_boundary(X, y, logreg)
```

# Train a basic logistic regression model and plot its decision boundary

```python
logreg = LogisticRegression(C=1e5)
logreg.fit(X, y)

def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.show()

plot_decision_boundary(X, y, logreg)
```

# Train an SVM model and plot its decision boundary plus margins

- Make the C parameter = 100.

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=100)
svm.fit(X, y)

def plot_decision_boundary_with_margin(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
    plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100,
                linewidth=1, facecolors='none', edgecolors='k')
    
    w = model.coef_[0]
    a = -w[0] / w[1]
    xx_ = np.linspace(x_min, x_max)
    yy_ = a * xx_ - (model.intercept_[0]) / w[1]
    plt.plot(xx_, yy_, 'k-')
    
    margin = 1 / np.sqrt(np.sum(model.coef_ ** 2))
    yy_down = yy_ - np.sqrt(1 + a ** 2) * margin
    yy_up = yy_ + np.sqrt(1 + a ** 2) * margin
    
    plt.plot(xx_, yy_down, 'k--')
    plt.plot(xx_, yy_up, 'k--')
    
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title('Decision Boundary and Margins - SVM')
    plt.show()

plot_decision_boundary_with_margin(X, y, svm)
```
